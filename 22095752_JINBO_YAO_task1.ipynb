Identify missing/NULL values and duplicate values in the datasets

```python
def find_nulls(data:pd.DataFrame) -> pd.DataFrame:
    temp = {'columns':[],'nulls':[],'nulls_%':[],'unique':[]}
    nulls = data.isnull().sum()
    temp['columns'] = nulls.axes[0]
    temp['nulls'] = nulls.values
    temp['nulls_%'] = nulls.values*100/data.shape[0]
    for i in nulls.axes[0]:
        temp['unique'].append(data[i].unique().shape[0])
    temp = pd.DataFrame(temp)
    return temp

find_nulls(df).style.background_gradient(cmap='Blues')
```

* For CompetitionDistance, choose -1 as the marker for missing values.
* For CompetitionOpenSinceMonth and CompetitionOpenSinceYear, use 0 or a special year such as 1900 to mark it.
* For Promo2SinceWeek, Promo2SinceYear and PromoInterval, if Promo2 is 0, it means that the second promotion is not carried out, so use the -1 mark

```python
import pandas as pd

# 假设df是你的数据框
# 使用-1标记CompetitionDistance的缺失值
df['CompetitionDistance'].fillna(-1, inplace=True)

# 使用1900标记CompetitionOpenSinceMonth和CompetitionOpenSinceYear的缺失值
df['CompetitionOpenSinceMonth'].fillna(0, inplace=True)
df['CompetitionOpenSinceYear'].fillna(1900, inplace=True)

# 对于Promo2SinceWeek、Promo2SinceYear和PromoInterval，如果Promo2为0，则填充-1
condition = df['Promo2'] == 0
df.loc[condition, 'Promo2SinceWeek'] = df.loc[condition, 'Promo2SinceWeek'].fillna(-1)
df.loc[condition, 'Promo2SinceYear'] = df.loc[condition, 'Promo2SinceYear'].fillna(-1)
df.loc[condition, 'PromoInterval'] = df.loc[condition, 'PromoInterval'].fillna('No Promo')

nulls = df.isnull().sum()
print(nulls)
```

# Data Pre-Processing & Exploratory Data Analysis (EDA)

The dataset is complete with no missing values or duplicate rows detected. While most columns are numerical, some of these represent categorical data, such as Store Type and State Holiday. Overall, the dataset encompasses a total of 4424 entries across 35 distinct columns.

To understand the nature of each variable with feature analysis, we can categorize them into numerical and categorical features based on the unique value counts for each feature:
1. Categorical Features: These are features with a relatively small number of unique values, which might represent categories.
2. Numerical Features: These features have a wide range of values and are likely to be continuous or discrete numerical data.

```python
numerical_variables = [
    "Customers",
    "CompetitionDistance",
    "CompetitionOpenSinceMonth",
    "CompetitionOpenSinceYear",
    "Promo2SinceYear",
    "Promo2SinceWeek"
]
categorical_variables = [
    "Store",
    "Open",
    "StateHoliday",
    "SchoolHoliday",
    "StoreType",
    "Assortment",
    "Promo",
    "Promo2",
    "PromoInterval"
]
num_variables = [
    "Sales",
    "Customers",
    "CompetitionDistance",
    "CompetitionOpenSinceMonth",
    "CompetitionOpenSinceYear",
    "Promo2SinceYear",
    "Promo2SinceWeek"
]
store_cnt = [
    "Open",
    "StateHoliday",
    "SchoolHoliday",
    "StoreType",
    "Assortment",
    "Promo",
    "Promo2",
    "PromoInterval"
]
# Analyzing the unique values in each column to determine if they are categorical or numerical
unique_values = df.nunique()
print(unique_values)
```

#### **Descriptive Statistics**

```python
import pandas as pd

results = {}

for feature in store_cnt:
    feature_sales = df.groupby([feature, 'Store'])['Sales'].sum().reset_index()
    highest_sales_stores = feature_sales.loc[feature_sales.groupby(feature)['Sales'].idxmax()]
    results[feature] = highest_sales_stores

# 打印结果
for feature, result in results.items():
    print(f"Highest sales stores for {feature}:")
    print(result)
    print("\n")

```

Chi-square Test

The p-values ​​for all variables are very low (close to 0), indicating statistical significance. This means that there is a significant correlation between these variables and the target variable, such as sales.

```python
import pandas as pd
from scipy.stats import chi2_contingency


results = []

for variable in categorical_variables:
    contingency_table = pd.crosstab(df[variable], df['Sales'])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    results.append((variable, chi2, p_value))

for result in results:
    print(f"Variable: {result[0]}, Chi-squared: {result[1]}, p-value: {result[2]}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_bar_chart(ax, data, title, xlabel, ylabel):
    sns.set(style="whitegrid")
    sns.barplot(ax=ax, x=data.index, y=data.values)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    # 在每个条上添加数值标签
    for index, value in enumerate(data):
        ax.text(index, value, f'{value}', ha='center', va='bottom')


# 根据属性数量动态创建子图
fig, axes = plt.subplots(2, len(store_cnt) // 2, figsize=(5 * (len(store_cnt) // 2), 8))

# 绘制每个属性的条形图
for i, attr in enumerate(store_cnt):
    row = i // (len(store_cnt) // 2)
    col = i % (len(store_cnt) // 2)
    counts = df[attr].value_counts()
    plot_bar_chart(axes[row, col], counts, f'Number of Stores by {attr}', attr, 'Number of Stores')

plt.tight_layout()
plt.show()
```
